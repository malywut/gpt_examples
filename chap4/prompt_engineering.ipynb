```python
def chat_completion(prompt, model='gpt-4', temperature=0):
    res = openai.ChatCompletion.create(
        model=model,
        messages=[{'role': 'user', 'content': prompt}],
        temperature=temperature
    )
    print(res['choices'][0]['message']['content'])
```
